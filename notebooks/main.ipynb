## 1.1 Differentiate the Categorical and Numerical columns

```python
# Dropping the unnecassary feature 'student_id'
df = df.drop("student_id", axis=1)
```

```python
cat_cols = df.select_dtypes('object')
num_cols = df.select_dtypes(['int64', 'float64'])
```

```python
cat_cols.info()
```

```python
num_cols.info()
```

```python
num_cols.hist(bins='auto', figsize=(20, 15))         # no cat because, hist requires 'number' of instances on X-axis and cat doesn't have any lol.
plt.show()
```

## 1.2 Listing the Cat and Num columns

```python
cat_cols_list = cat_cols.columns.to_list()
num_cols_list = num_cols.columns.to_list()

print("Categorical Columns: ", cat_cols_list)
print("\nNumerical Columns: ", num_cols_list)
```

## 1.3 Categorical columns profiling

```python
# Categorical column analysis
for col in cat_cols_list:
    print(f"\n{'='*30}\nColumn: {col}")
    print("Unique values:", df[col].nunique())
    print("\nValue Counts:\n", df[col].value_counts(dropna=False))
```

## 1.4 Correlation Matrix for Num Features

```python
corr_matrix = num_cols.corr()
```

```python
# comparing the correlation matrix of all the features with the target feature.
corr_matrix["exam_score"].sort_values(ascending=False)
```

## 1.5 Plotting the correlation matrix

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=.5)
plt.title("Correlation Matrix - Numerical Features")

plt.show()
```

## 1.6 Visual Exploration – Univariate, Bivariate, Outlier Checks

### Univariate Analysis of Each Num Feature

```python
for col in num_cols_list:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("counts")
    plt.tight_layout()
    plt.show()
```

```python
for col in cat_cols_list:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f"Category Distribution: {col}")
    plt.xticks(rotation=45)
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.show()
```
