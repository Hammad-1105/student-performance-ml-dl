## 2.1 Encoding Categorical Features

```python
cat_cols.head()
```

```python
cat_train = X_train[cat_cols_list].copy()
```

```python
cat_bin = ["part_time_job", "extracurricular_participation", "dropout_risk", "access_to_tutoring"]
cat_ordinal = ["diet_quality", "internet_quality", "parental_education_level", "family_income_range"]
cat_nominal = [c for c in cat_cols_list if c not in cat_bin + cat_ordinal]
```

```python
# quick sanity
print("binary:", cat_bin) 
print("ordinal:", cat_ordinal) # ordinal encoding
print("nominal:", cat_nominal) # onehot encoding
```

### Step 2.1.1 – Binary Encoding (Label Encoding)

```python
bin_map = {"No": 0, "Yes": 1}

cat_bin_df = cat_train[cat_bin].copy()

for c in cat_bin:
    cat_bin_df[c] = cat_bin_df[c].map(bin_map)
```

```python
cat_bin_df.head()
```

### Step 2.1.2 - Ordinal Encoding 

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the correct category order for each ordinal column
categories_order = [
    ['Poor', 'Fair', 'Good'],  # diet_quality → Poor < Fair < Good
    ['Low', 'Medium', 'High'],  # internet_quality → Low < Medium < High
    ['High School', 'Some College', 'Bachelor', 'Master', 'PhD'],  # parental_education_level
    ['Low', 'Medium', 'High']  # family_income_range
]

ordinal_encoder = OrdinalEncoder(categories=categories_order)
```

```python
cat_ordinal_df = cat_train[cat_ordinal].copy()
cat_ordinal_df.head()
```

```python
cat_ordinal_df = ordinal_encoder.fit_transform(cat_ordinal_df)
```

```python
cat_ordinal_df = pd.DataFrame(cat_ordinal_df, columns=cat_ordinal, index=cat_train[cat_ordinal].index)
```

```python
cat_ordinal_df.head()
```

```python
for col_name, cats in zip(cat_ordinal, ordinal_encoder.categories_):
    print(f"{col_name} → {list(cats)}")
```

### Step 2.1.3 - Nominal Encoding 

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
cat_nominal_df = cat_train[cat_nominal].copy()
cat_nominal_df.head()
```

```python
cat_nominal_encoded = onehot_encoder.fit_transform(cat_nominal_df)
```

```python
# Get category names for each column
nominal_column_names = []
for col, categories in zip(cat_nominal, onehot_encoder.categories_):
    nominal_column_names.extend([f"{cat} ({col})" for cat in categories])
```

```python
# Create DataFrame with proper column names & original index
cat_nominal_df = pd.DataFrame(cat_nominal_encoded, columns=nominal_column_names, index=cat_train.index)
```

```python
cat_nominal_df.head()
```

### Encoding finished, finalizing the train set

```python
cat_encoded_train = pd.concat([cat_bin_df, cat_ordinal_df, cat_nominal_df], axis=1)
```

## 2.4 Final Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

### Defining the feature groups

```python
# Binary categorical (already 0/1, so passthrough)
cat_bin = ['part_time_job', 'extracurricular_participation', 
           'dropout_risk', 'access_to_tutoring']

# Ordinal categorical (with defined order)
cat_ordinal = ['diet_quality', 'internet_quality', 
               'parental_education_level', 'family_income_range']

# Nominal categorical (no order)
cat_nominal = ['gender', 'major', 'study_environment', 'learning_style']

# Numerical
num_features = ['age', 'study_hours_per_day', 'social_media_hours', 
                'netflix_hours', 'attendance_percentage', 'sleep_hours', 
                'exercise_frequency', 'mental_health_rating', 
                'previous_gpa', 'semester', 'stress_level', 
                'social_activity', 'screen_time', 
                'parental_support_level', 'motivation_level', 
                'exam_anxiety_score', 'time_management_score']
```

### Defining the pipelines

```python
# Numerical pipeline
num_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Ordinal Pipeline
ordinal_pipeline = Pipeline([
    ('ordinal', OrdinalEncoder(categories=[
        ['Poor', 'Fair', 'Good'],                # diet_quality
        ['Low', 'Medium', 'High'],               # internet_quality
        ['High School', 'Some College', 'Bachelor', 'Master', 'PhD'],  # parental_education_level
        ['Low', 'Medium', 'High']                # family_income_range
    ]))
])

# Nominal Pipeline
nominal_pipeline = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Binary Pipeline - FIXED: Need categories for all 4 binary features
binary_pipeline = Pipeline([
    ('bin_encoder', OrdinalEncoder(categories=[
        ['No', 'Yes'],  # part_time_job
        ['No', 'Yes'],  # extracurricular_participation
        ['No', 'Yes'],  # dropout_risk
        ['No', 'Yes']   # access_to_tutoring
    ]))
])
```

### Defining the Column Transformer

```python
# Column Transformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('ord', ordinal_pipeline, cat_ordinal),
    ('nom', nominal_pipeline, cat_nominal),
    ('bin', binary_pipeline, cat_bin)
])
```

### Transforming train and test set

```python
X_train_prepared = preprocessor.fit_transform(X_train)  
print("Train shape:", X_train_prepared.shape)
```
